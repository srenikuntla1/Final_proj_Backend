```python
from geopy.distance import geodesic
# Assuming 'schools_df' has columns 'X' (longitude) and 'Y' (latitude) for schools
# And 'block_groups_df' has a 'centroid' column with Point geometries for block group centroids

# Convert schools_df to a GeoDataFrame with Point geometries
schools_gdf = gpd.GeoDataFrame(
    schools_df,
    geometry=gpd.points_from_xy(schools_df.X, schools_df.Y),
    crs="EPSG:4326"
)

# Ensure block_groups_df is a GeoDataFrame with the correct geometry
block_groups_gdf = gpd.GeoDataFrame(
    block_groups_df,
    geometry=block_groups_df['centroid'].apply(lambda x: Point(x.x, x.y)),
    crs="EPSG:4326"
)

# Create a DataFrame to store the distances
distances_df = pd.DataFrame(index=block_groups_gdf.index, columns=schools_gdf.index)

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in miles between two points 
    on the earth (specified in decimal degrees)
    """
    # Convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    # Haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    r = 3956 # Radius of Earth in miles. Use 6371 for kilometers
    return c * r

# Replace the calculate_distance_miles function with haversine
for bg_index, bg_row in block_groups_gdf.iterrows():
    for school_index, school_row in schools_gdf.iterrows():
        distances_df.at[bg_index, school_index] = haversine(bg_row['geometry'].x, bg_row['geometry'].y, school_row['geometry'].x, school_row['geometry'].y)
# distances_df now contains the distances in miles from each block group to each school
print(distances_df)

# the latitude and the logitude of the schools are represented by the columns X and Y respectively
# the latitude and the logitude of the block groups are represented by the columns INTPTLAT and INTPTLON respectively

# block_groups_df['nearest_school'] = None

# block_groups_df['distance_miles'] = None

# for i, block_group in block_groups_df.iterrows():
#     min_distance = float('inf')
#     nearest_school = None
#     for j, school in Geocoded_Georgia_high_schools_df.iterrows():
#         distance = geopy.distance.distance((block_group['centroid'].y, block_group['centroid'].x), (school['Y'], school['X'])).miles
#         if distance < min_distance:
#             min_distance = distance
#             nearest_school = school['USER_Sch_2']
#     block_groups_df.at[i, 'nearest_school'] = nearest_school
#     block_groups_df.at[i, 'distance_miles'] = min_distance

# block_groups_df

# # Assuming 'centroid' and 'Geometry' are the geometry columns for block_groups_df and Geocoded_Georgia_high_schools_df respectively
# block_groups_df = gpd.GeoDataFrame(block_groups_df, geometry='centroid')
# Geocoded_Georgia_high_schools_df = gpd.GeoDataFrame(Geocoded_Georgia_high_schools_df, geometry='geometry')

# # Now you can safely call to_crs and other geospatial methods
# block_groups_df = block_groups_df.to_crs(Geocoded_Georgia_high_schools_df.crs)

# # Step 1: Ensure CRS match
# block_groups_df = block_groups_df.to_crs(Geocoded_Georgia_high_schools_df.crs)

# # make the crs of the block_groups_df and Geocoded_Georgia_high_schools_df same
# block_groups_df.crs = Geocoded_Georgia_high_schools_df.crs




# # Step 2: Calculate distances and find nearest school
# def find_nearest_school(block_group_row):
#     # Temporary list to store distances and school names
#     distances_schools = []
    
#     # Calculate distance to each school
#     for _, school_row in Geocoded_Georgia_high_schools_df.iterrows():
#         # Calculate distance
#         distance = block_group_row['centroid'].distance(school_row['geometry'])
#         # Convert distance to miles if needed, assuming CRS is in meters
#         distance_miles = distance * 0.000621371
#         distances_schools.append((school_row['School Name'], distance_miles))
    
#     # Find the nearest school
#     nearest_school, nearest_distance = min(distances_schools, key=lambda x: x[1])
    
#     return pd.Series([nearest_school, nearest_distance], index=['nearest_school', 'distance_miles'])

# # Apply the function to each row in block_groups_df
# block_groups_df[['nearest_school', 'distance_miles']] = block_groups_df.apply(find_nearest_school, axis=1)
```
